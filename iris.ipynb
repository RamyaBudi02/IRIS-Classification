```python
x=df.iloc[: , :4]#to seperate depenedent and independent features
y=df.iloc[:, 4]
print("X:\n...........\n{}".format(x))
print("X:\n...........\n{}".format(y))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=0)
print("X_train:\n{}".format(X_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)
print(y_pred)#machine predicted target
print("Y_test:\n{}".format(Y_test))#original target
```

```python
#as the above output is difficult to compare
#we use confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(Y_test,y_pred))
accuracy=accuracy_score(Y_test,y_pred)*100
print("accuracy of the model is {:.2f}".format(accuracy))
```

```python
#finding the target name for the given feature
X_new=np.array([[151,2.3,2.9,5.0]])
print("X_new.shape:{}".format(X_new.shape))
y_new=knn.predict(X_new)
print("prediction \n{}".format(y_new))
```
